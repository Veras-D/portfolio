```python
import pandas as pd
from ydata_profiling import ProfileReport

report = pd.read_csv("Titanic-Dataset.csv")
report = report.drop(["PassengerId", "Cabin", "Name", "Ticket"], axis=1)
report.loc[report['Age'] < 1, 'Age'] = report['Age'] * 100
report.loc[report['Age'] == 1, 'Age'] = 10
report["Age"].fillna(report["Age"].median(), inplace=True)
report["Age"] = report["Age"].astype(int)
report = report.dropna()
report.info()
```

```python
report.boxplot(column="Age")
```

```python
profile = ProfileReport(report, title="Titanic")
profile.to_notebook_iframe()
# profile.to_file("books_data.html")
```
