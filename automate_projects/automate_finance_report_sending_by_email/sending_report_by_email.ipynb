## Selecting the Best Stocks
- Stocks with P/E Ratio > 0
- Stocks with Dividend Yield > 5%
- Net Profit Margin > 30%


```python
import fundamentus

df = fundamentus.get_resultado()
filtro = df[(df.pl > 0) & (df.dy > 0.05) & (df.mrgebit > 0.3)]
filtro.sort_values('dy', ascending = False, inplace = True)
filtro
```
